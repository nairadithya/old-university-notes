```python
import pandas as pd

df = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')
```

```python
print(df.info())
# No Missing values found in the datset
```

```python
# Outliers

print(df.shape)

Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

l = Q1 - 1.5 * IQR
u = Q3 + 1.5 * IQR

df = df[(df['Age'] >= l) & (df['Age'] <= u)]

print(df.shape)
```

```python
# Duplicates

duplicate_rows = df[df.duplicated()]
print("Number of Duplicate Rows = " , duplicate_rows.shape)

df = df.drop_duplicates()
print("Size of the Dataset after removing the duplicates = " , df.shape)
```

```python
# Normalisation 
range_of_columns = df.max() - df.min()
print(range_of_columns)
# Since the scale is not as huge as to affect the learning process of an algorithm, it's better not to do it.
```

```python
# One hot encoding on GenHlth, Income and Education
df = pd.concat([df,pd.get_dummies(df['GenHlth'],prefix = str('GenHlth'),prefix_sep='_')],axis = 1)
df.drop('GenHlth',axis=1)

df = pd.concat([df,pd.get_dummies(df['Education'],prefix = str('Education'),prefix_sep='_')],axis = 1)
df.drop('Education',axis=1)

df = pd.concat([df,pd.get_dummies(df['Income'],prefix = str('Income'),prefix_sep='_')],axis = 1)
df.drop('Income',axis=1)
```

```python
print(df.info())
```

```python
# Outliers
# univariate analysis

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(df['BMI'])
plt.show()
print(df.shape)

Q1 = df['BMI'].quantile(0.25)
Q3 = df['BMI'].quantile(0.75)
IQR = Q3 - Q1

l = Q1 - 1.5 * IQR
u = Q3 + 1.5 * IQR

df = df[(df['BMI'] >= l) & (df['BMI'] <= u)]

sns.boxplot(df['BMI'])
plt.show()
print(df.shape)
```

```python
import matplotlib.pyplot as plt

# univariate analysis
# Histogram
df['BMI'].hist(bins=20)
plt.show()
```

```python

x = df.drop(['Diabetes_012'], axis=1)
y = df['Diabetes_012']

# class imbalance
value_counts = df['Diabetes_012'].value_counts()
print("Unique values and their frequencies in 'target' : ")
print(value_counts)
class_counts = df['Diabetes_012'].value_counts()

# Pie chart
# univariate analysis
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Diabetes Class Distribution')
plt.axis('equal')  
plt.show()

# Doing EDA and the pie chart shows that there is a high imbalance in the data
```

```python
# bivariate analysis
# Correlation Analysis

d = df[['BMI', 'Age', 'MentHlth', 'PhysHlth']]

# Correlation matrix 
corr = d.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

#This matrix shows that the features 'Age' and 'PhysHlth' are in the same direction
```
